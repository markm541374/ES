```python
#standard library imports and some useful functions

import scipy as sp
from scipy import linalg as spl
from matplotlib import pyplot as plt
import time
%matplotlib inline

#given a 1d input return a matrix with the input on the diagonal
def vec2trace(v):
    (l,_) = v.shape
    M = sp.eye(l)
    for i in range(l):
        M[i,i]=v[i]
    return M

# return xlogx but return zero for x=0 instead of raising a log0 error
def xlx0(x):
    if x==0:
        return 0
    else:
        return x*sp.log(x)
    
#sqexp kernel generator
def gen_sqexp_k(theta):
	A=theta[0]
	d=sp.matrix(theta[1:]).T
	
	N=sp.matrix(d).shape[0]
	D=sp.eye(N)
	for i in range(N):
		D[i,i]=1./(d[i,0]**2)
	return lambda x,y:A*sp.exp(-0.5*(x-y)*D*(x-y).T)

#sqexp kernel generator
def sqexp_k_d(theta,x1,x2,d1=sp.NaN,d2=sp.NaN):
    
    x1=sp.matrix(x1)
    x2=sp.matrix(x2)
    A=theta[0]
    d=sp.matrix(theta[1:]).T
    N=sp.matrix(d).shape[0]
    D=sp.eye(N)
    for i in range(N):
        D[i,i]=1./(d[i,0]**2)
    cov_core=A*sp.exp(-0.5*(x1-x2)*D*(x1-x2).T)
    if sp.isnan(d1) and sp.isnan(d2):
        cc= cov_core
    elif sp.isnan(d1):
        d2=int(d2)
        delta_axis=x2[d2,0]-x1[d2,0]
    
        cc= -D[d2,d2]*delta_axis*cov_core
    elif sp.isnan(d2):
        d1=int(d1)
        delta_axis=x1[d1,0]-x2[d1,0]
        
        cc= -D[d1,d1]*delta_axis*cov_core
    elif int(d1)==int(d2):
        d=int(d1)
        th = D[d1,d1]
        cc= th*(1-th*(x1[d,0]-x2[d,0])**2)*cov_core
    else:
        d1=int(d1)
        d2=int(d2)
        cc= D[d2,d2]*D[d1,d1]*(x1[d1,0]-x2[d1,0])*(x1[d2,0]-x2[d2,0])*cov_core
    
    return cc[0,0]
from functools import partial

def gen_sqexp_k_d(theta):
    k = partial(sqexp_k_d,theta)
    return k


#k builder for asymettric k
def buildKasym_d(kf,x1,x2,d1,d2):
        #x should be column vectors, returns a matrix with x1 rows, x2 columns
        (l1,_)=x1.shape
        (l2,_)=x2.shape
        K=sp.matrix(sp.empty([l1,l2]))
        for i in range(l1):
            for j in range(l2):
                K[i,j]=kf(x1[i,:],x2[j,:],d1=d1[i,0],d2=d2[j,0])
                
        return K
# builder for symmetric k. 10**-10 is added tot eh diagonal for numerical stabulity

# builder for symmetric k. 10**-10 is added tot eh diagonal for numerical stabulity
def buildKsym_d(kf,x,d):
        #x should be  column vector
        (l,_)=x.shape
        K=sp.matrix(sp.empty([l,l]))
        for i in range(l):
            K[i,i]=kf(x[i,:],x[i,:],d1=d[i,0],d2=d[i,0])+10**-10
            for j in range(i+1,l):
                K[i,j]=kf(x[i,:],x[j,:],d1=d[i,0],d2=d[j,0])
                
                K[j,i]=K[i,j]
                
        return K
    
class GPcore:
    def __init__(self,X_s,Y_s,S_s,D_s,kf):
        #print "GPcore init"
        #print X_s.T
        #print Y_s.T
        #print S_s.T
        self.D_s=D_s
        self.X_s=X_s
        #self.Y_s=Y_s
        #self.S_s=S_s
        self.kf=kf
        
        K_ss = buildKsym_d(kf,X_s,D_s)
        
        self.K_ss_cf = spl.cho_factor(K_ss+vec2trace(S_s))
        self.a = spl.cho_solve(self.K_ss_cf,Y_s)
        
        #print "----------"
        return
    
    def changeY(self,Y_s):
        #print "changeY"
        #print Y_s.T
        
        self.a = spl.cho_solve(self.K_ss_cf,Y_s)
        #print "a"
        #print self.a.T
        return
    
    def infer_m(self,X_i,D_i):
        K_si = buildKasym_d(kf,X_i,self.X_s,D_i,self.D_s)
        m = K_si*self.a
        return m
    
    def infer_full(self,X_i,D_i):
        K_ii = buildKsym_d(kf,X_i,D_i)
        K_si = buildKasym_d(kf,X_i,self.X_s,D_i,self.D_s)
        
        m = K_si*self.a
        V = K_ii - K_si*spl.cho_solve(self.K_ss_cf,K_si.T)
        
        return [m,V]
    
    def infer_diag(self,X_i,D_i):
        K_si = buildKasym_d(kf,X_i,self.X_s,D_i,self.D_s)
        
        m = K_si*self.a
        V=sp.matrix(sp.zeros(X_i.shape[0])).T
        for j,r in enumerate(K_si):
            
            V[j,0]=self.kf(X_i[j,0],X_i[j,0],d1=D_i[j,0],d2=D_i[j,0])-r*spl.cho_solve(self.K_ss_cf,r.T)
        return [m,V]
kf = gen_sqexp_k_d([1.,0.3])
```

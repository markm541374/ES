```python
kf = GPep.gen_sqexp_k_d([1.,0.3])

Xo=sp.matrix([[-0.85],[-0.7],[0.1],[0.4]])
Yo=sp.matrix([[0.35],[0.2],[-0.4],[-0.1]])
So=sp.matrix([[0.01],[0.01],[0.01],[0.01]])
Do=[[sp.NaN],[sp.NaN],[sp.NaN],[sp.NaN]]

```

```python
g1=GPd.GPcore(Xo,Yo,So,Do,kf)
a=GPd.plot1(g1,[-1],[1])

#make a draw on a uniform dist
X_i = sp.matrix(sp.linspace(-1,1,200)).T
D_i = [[sp.NaN]]*200
mi,Vi=g1.infer_full(X_i,D_i)
Vi_cho = spl.cholesky(Vi,lower=True)
dr = mi+Vi_cho*sp.matrix(sp.random.normal(size=200)).T
a.plot(sp.array(X_i).flatten(),sp.array(dr).flatten(),'g')
xsi=dr.argmin()
xs=X_i[xsi,:][0,0]
print xs
a.plot(xs,dr[xsi,0],'go')
```

```python

nh=101
ns=30

X_h = sp.matrix(sp.linspace(-1,1,nh)).T
D_h = [[sp.NaN]]*nh
mh,Vh=g1.infer_full(X_h,D_h)
Vh_cho = spl.cholesky(Vh,lower=True)

H=sp.zeros([nh,ns])
ss=sp.logspace(4,-4,ns)

nd=400
for i in xrange(nd):
    if i%25==0:
        print i
    dr = mh+Vh_cho*sp.matrix(sp.random.normal(size=nh)).T
    xsi=dr.argmin()
    xs=X_h[xsi,:][0,0]
    #eq constraint
    
    Xg=sp.matrix([[xs]])
    Yg=sp.matrix([[0.]])
    Sg=sp.matrix([[0.000000001]])
    Dg=[[0]]
    
    [Xc,Yc,Sc,Dc]=GPep.catObs([[Xo,Yo,So,Do],[Xg,Yg,Sg,Dg]])
    
    #ineq constraints
    
    Xz=sp.matrix([[xs],[xs]])
    #Sz=sp.matrix([[10**-9],[10**-9]])
    Dz=[[sp.NaN],[0,0]]
    #the inequality
    Iz=sp.matrix([[Yo[Yo.argmin(),:][0,0]],[0.]])
    #sign of the inequality
    Gz=[0,0]
    Nz=[0.1,0.]
    
    g=GPep.GPcore(Xc,Yc,Sc,Dc,Xz,Dz,Iz,Gz,Nz,kf)
    #GPep.plot1(g,[-1],[1])
    for j in xrange(nh):
        Xt=sp.matrix([[X_h[j,0]],[xs]])
        Dt=[[sp.NaN],[sp.NaN]]
        m,V=g.infer_full(Xt,Dt)
        s=V[0,0]+V[1,1]-V[0,1]-V[1,0]
        mu=m[1,0]-m[0,0]
        alpha=mu/sp.sqrt(s)
        beta=sps.norm.pdf(alpha)/sps.norm.cdf(alpha)
        
        vnxxs=V[0,0]-beta*(beta+alpha)*(1./s)*(V[0,0]-V[0,1])**2
        for k in xrange(ns):
            Hydxxs=0.5*sp.log(2*sp.pi*sp.e*(vnxxs+ss[k]))
            H[j,k]+=Hydxxs
H=H/float(nd)
#Ht=sp.zeros([nh,ns])
for i in xrange(nh):
    m,v=g1.infer_diag(sp.matrix([[X_h[i,0]]]),[[sp.NaN]])
    for k in xrange(ns):
        Hydx=0.5*sp.log(2*sp.pi*sp.e*(v[0,0]+ss[k]))
        H[i,k]=Hydx-H[i,k]
```

```python
a=GPd.plot1(g1,[-1],[1])
for k in range(ns):
    a.plot(sp.array(X_h).flatten(),H[:,k],'r')
```

```python
bs=[0.01+sp.sqrt(1./float(i)) for i in ss]

U=sp.zeros([nh,ns])
plt.figure(figsize=(10,10))
for i in xrange(nh):
    U[i,:]=[H[i,j]/float(bs[j]) for j in xrange(ns)]
    plt.semilogx(bs,U[i,:])
    #plt.semilogx(bs,H[i,:])
```

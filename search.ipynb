```python
from scipy import interpolate
class fgen1d():
    def __init__(self,lower,upper,npoints,kf):
        self.npoints=npoints
        self.Xi=sp.matrix(sp.linspace(lower,upper,npoints)).T
        V_p=GPd.buildKsym_d(kf,self.Xi,[[sp.NaN]]*npoints)
        self.vpCho=spl.cholesky(V_p,lower=True)
        print 'done'
        return
    def genfun(self):
        fn = self.vpCho*sp.matrix(sp.random.normal(size=self.npoints)).T
        f = sp.interpolate.interp1d(sp.array(self.Xi).flatten(),sp.array(fn).flatten())
        return f
kf = GPep.gen_sqexp_k_d([1.,0.2])
ff=fgen1d(-1,1,1001,kf)
```

```python
f=ff.genfun()
plt.plot(sp.linspace(-1,1,100),map(f,sp.linspace(-1,1,100)))
#print ff.vpCho.shape
```

```python
n_init=5
x=sp.random.uniform(-1,1,n_init)
y=map(f,x)+sp.random.normal(scale=0.1,size=n_init)
Xo=sp.matrix(x).T
Yo=sp.matrix(y).T
So=sp.matrix([[0.01]]*n_init)
Do=[[sp.NaN]]*n_init
g1=GPd.GPcore(Xo,Yo,So,Do,kf)
a=GPd.plot1(g1,[-1],[1])
```

```python
Xi=sp.linspace(-1,1,401)
#Xi=[0,12]
Si=[0.01]
H=EntropyPredict.predict_1dfixk([Xo,Yo,So,Do],kf,Xi,Si,nd=50)
EI=EntropyPredict.EI_1dfixk([Xo,Yo,So,Do],kf,Xi,sp.amin(Yo))
```

```python
a=GPd.plot1(g1,[-1],[1])
a.plot(Xi,H[:,0],'r')
a.plot(Xi,3*EI,'g')
a.plot(Xi[sp.argmax(EI)],3*max(EI),'go')
```
